```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
AGUSData = pd.read_csv("/Users/zhangxiangyu/Desktop/RA/Agt Vs. US/ARG VS. US.csv")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for country, data in AGUSData.groupby('country'):
    ax.plot(data['year'], data['pop'], label=country)

ax.set_xlabel('Year')
ax.set_ylabel('Population (in thousands)')
ax.set_title('Population Growth for Each Country from 1924 to 1938')
ax.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for country, data in AGUSData.groupby('country'):
    ax.plot(data['year'], data['rgdp'], label=country)

ax.set_xlabel('Year')
ax.set_ylabel('Real GDP in 2011$(in thousands)')
ax.set_title('Real GDP for Each Country from 1924 to 1938')
ax.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for country, data in AGUSData.groupby('country'):
    ax.plot(data['year'], data['gdppc'], label=country)

ax.set_xlabel('Year')
ax.set_ylabel('Real GDP per capita in 2011$')
ax.set_title('Real GDP per capita for Each Country from 1924 to 1938')
ax.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for country, data in AGUSData.groupby('country'):
    ax.plot(data['year'], data['rgdpgrowth'], label=country)

ax.set_xlabel('Year')
ax.set_ylabel('Real GDP growth (%)')
ax.set_title('Real GDP growth for Each Country from 1924 to 1938')
ax.legend()
plt.grid(True)
plt.show()
```
